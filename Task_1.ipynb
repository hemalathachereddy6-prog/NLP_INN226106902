Task 1: Conceptual Understanding

1. Difference between "Love" and "love" in NLP

In NLP, "Love" (capital L) and "love" (small l) are treated as different tokens if the text is not normalized.

"Love" may appear at the beginning of a sentence or as a proper noun (like a name).
"love" is usually a common word representing emotion.

If we don’t apply case normalization (lowercasing), the model will think they are two separate words, which increases vocabulary size unnecessarily and may reduce accuracy.

 So, NLP systems often convert everything to lowercase to treat them as the same word.

2. What happens if stopwords are not removed?

If stopwords (like is, the, and, in) are not removed:

The dataset becomes noisy and large
Important words get less importance
Model training becomes slower
Accuracy may decrease because common words dominate

 However, keeping stopwords may sometimes be useful depending on the task (like sentiment analysis).

3. Two real-world scenarios where removing stopwords can be harmful

(a) Sentiment Analysis

Example:

"This movie is not good"

If we remove stopwords like "not", the sentence becomes:

"This movie good"

 This completely changes the meaning from negative to positive.

(b) Question Answering / Search Systems

Example:

"What is the best college in Bangalore?"

If stopwords are removed:

"best college Bangalore"

 The system may lose the intent of the question and return less accurate results.

4. Difference between Stemming and Lemmatization
Feature	Stemming	Lemmatization
Meaning	Cuts words to root form	Converts words to meaningful base form
Output	May not be a real word	Always a valid word
Method	Rule-based chopping	Uses vocabulary & grammar
Example	"running" → "run" or "runn"	"running" → "run"
Accuracy	Less accurate	More accurate
Speed	Faster	Slower

 In simple terms:

Stemming = rough cutting
Lemmatization = smart, meaningful conversion






Task 2:Build Advanced Preprocessing Function


```python

import re

def preprocess_text(text):
    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # 3. Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters (soooo → soo, goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # 6. Remove special characters / punctuation
    text = re.sub(r'[^a-z\s]', '', text)

    # 7. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 8. Remove very short tokens (length ≤ 2), keep "no", "not"
    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    # Join back
    cleaned_text = ' '.join(filtered_words)

    return cleaned_text
```

Task 6: Build Full Pipeline

```python



import re


# Preprocessing Function

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return filtered_words



# Full Pipeline Function

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }



# Example Usage

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Visit https://openai.com now!"
]

result = full_pipeline(sentences)

print("Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```

Task 7: Error Handling


```python
import re


# Safe Preprocessing Function

def preprocess_text(text):

    # Handle empty or invalid input
    if not text or not isinstance(text, str):
        return []

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    # If text becomes empty after cleaning
    if text == "":
        return []

    words = text.split()
    filtered_words = [word for word in words if len(word) > 2 or word in ['no', 'not']]

    return filtered_words



# Full Pipeline with Error Handling

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)

        # Handle cases where no tokens remain
        if not tokens:
            clean_sentence = ""
        else:
            clean_sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Test Cases (Error Handling)

test_inputs = [
    "",                     # Empty string
    "😍😂🔥",               # Only emojis
    "1234567890",           # Only numbers
    "Normal text example"   # Valid input
]

result = full_pipeline(test_inputs)

print("Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```
